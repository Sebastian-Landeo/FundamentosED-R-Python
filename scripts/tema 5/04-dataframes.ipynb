# Manipulacin de data frames

```python
df = pd.DataFrame({
    "key": ["A", "B", "C", "A", "B", "C"],
    "value1": range(6),
    "value2": np.random.randint(0, 100, 6)
})
```

```python
df
```

```python
df.groupby("key").aggregate(["min", "median", "mean", "max"])
```

```python
df.groupby("key").aggregate({
    "value1": "min",
    "value2": "max"
})
```

## Filter

```python
df.groupby("key").std()
```

```python
def filter_func(x):
    return x["value2"].std()>7.5
```

```python
df.groupby("key").filter(filter_func)
```

```python
df["value2"]
```

```python
df.value2
```

## Transform

```python
df.groupby("key").transform(lambda x: x - x.mean())
```

## Apply

```python
def norm_by_col2(x):
    x["value1"]=x["value1"]/x["value2"].sum()
    return x
```

```python
df.groupby("key").apply(norm_by_col2)
```

## Divisin de datos por columnas

```python
L = [0, 1, 0, 1, 2, 0]
print(df)
df.groupby(L).sum()
```

```python
df.groupby("key").sum()
```

```python
df.groupby(df["key"]).sum()
```

```python
df2 = df.set_index("key")
df2
```

```python
mapping = {
    "A": "vowel",
    "B": "consonant",
    "C": "consonant"
}
```

```python
df2.groupby(mapping).sum()
```

```python
df2.groupby(str.lower).mean()
```

```python
df2.groupby([str.lower, mapping]).mean()
```

## Grficos desde data frames y exportacin

```python
import numpy as np
years = [year for year in range(1900, 2020)]
deads = [(y + np.random.uniform(0, 100)-1850) for y in years]
```

```python
df = pd.DataFrame({
    "year": years,
    "deads": deads
})
```

```python
df.head()
```

```python
df.plot(kind = "scatter", x="year", y="deads", figsize = (10, 8))
```

```python
my_plot = df.plot(x="year", y="deads", figsize = (10, 8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("../tema 5/data_frame_plot.png")
```
