# Manipulación de data frames

```python
df = pd.DataFrame({
    "key": ["A", "B", "C", "A", "B", "C"],
    "value1": range(6),
    "value2": np.random.randint(0, 100, 6)
})
```

```python
df
```

```python
df.groupby("key").aggregate(["min", "median", "mean", "max"])
```

```python
df.groupby("key").aggregate({
    "value1": "min",
    "value2": "max"
})
```

## Filter

```python
df.groupby("key").std()
```

```python
def filter_func(x):
    return x["value2"].std()>7.5
```

```python
df.groupby("key").filter(filter_func)
```

```python
df["value2"]
```

```python
df.value2
```

## Transform

```python
df.groupby("key").transform(lambda x: x - x.mean())
```

## Apply

```python
def norm_by_col2(x):
    x["value1"]=x["value1"]/x["value2"].sum()
    return x
```

```python
df.groupby("key").apply(norm_by_col2)
```

### Pregunta 3

Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio y
compara el valor de los que hacen ejercicio habitualmente y los que no. ¿Observas mucha diferencia?

- En esta versión se agrega nuevas columnas

```python
# Calcular la variación del pulso
dft['variacion_pulso'] = dft['pulso.despues'] - dft['pulso.antes']
# Calcular el porcentaje de variación
dft['porcentaje_variacion'] = (dft['variacion_pulso'] / dft['pulso.antes']) * 100
# Agrupar por si hace deporte o no
mean_variacion = dft.groupby('hace.deporte')['porcentaje_variacion'].mean()
print(mean_variacion)
```

```python
dft
```

- En esta versión se usa el **apply**

```python
# Calcular la variación del pulso y el porcentaje de variación usando apply y groupby
def calcular_porcentaje_variacion(grupo):
    variacion_pulso = grupo['pulso.despues'] - grupo['pulso.antes']
    porcentaje_variacion = (variacion_pulso / grupo['pulso.antes']) * 100
    return porcentaje_variacion.mean()

dft.groupby('hace.deporte').apply(calcular_porcentaje_variacion)
```

```python
dft.groupby(['hace.deporte', 'genero']).apply(calcular_porcentaje_variacion)
```

```python
dft[dft['hace.deporte']=="si"].groupby("genero").apply(calcular_porcentaje_variacion)
```

```python
prueba = dft.groupby("genero").apply(calcular_porcentaje_variacion).filter(dft['hace.deporte']=="si")
print(prueba)
```

El problema con el código es que estás intentando aplicar un filtro en una serie que no es adecuada para esta operación directamente después de usar groupby y apply. La forma correcta sería filtrar los datos antes de agruparlos y aplicar la función.

```python
# dft.filter('hace.deporte'=="si").groupby("genero").apply(calcular_porcentaje_variacion)
# Tampoco funciona
dft.filter(items=["genero"])
```
